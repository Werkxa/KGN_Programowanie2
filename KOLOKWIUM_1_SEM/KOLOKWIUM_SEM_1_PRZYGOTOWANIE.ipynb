```python
# 2 Funkcja max_of_three(a, b, c) zwracająca największą z trzech liczb. (instrukcje warunkowe)

# def max_of_three(a, b, c):
#   if a > b | c:
#     return a
#   elif b > a |c:
#     return b
#   else:
#     return c


def max_of_three(a, b, c):
  l = [a,b,c]
  counter = l[0] #zakładamy, że pierwszy el jest największy;
  #zmienna pomocnicza- pierwsza wartość ze zbioru (gdy byłoby tam 0, w przypadku)
  #liczb ujemnych, program błędnie by podał, że 0 jest największe
  for elem in l:
    if elem > counter: #"Czy ta liczba jest większa od naszego lidera?"
      counter = elem #jeśli tak, to ona zostaje nowym liderem- podmieniamy wartość
  return counter


max_of_three(-3,8,8)
```

```python
#2.1 Najmniejsza z trzech

def min_of_three(a,b,c):
  l = [a,b,c]
  counter = l[0]
  for elem in l:
    if elem < counter: #zmieniamy większe > na mniejsze <
      counter = elem
  return counter

min_of_three(2,7,-11)
```

```python
# 4 Funkcja sum_list(lst) zwracająca sumę elementów listy liczb. (listy, pętla for); MOZE SIE POJAWIC LICZBA EL DLUGOSC LISTY ELEM TO JEST 1

def sum_list(lst):
  counter = 0 #skarbiec- zaczynamy od zera; gdyby 1 to zawsze o jeden za duzy wynik
  for elem in lst: #"nie obcholdzi mnie ile tych elem jest, wez po kolei kazdy z nich i nazwij go tymczasowo elem"
    counter += elem #za kazdym razem gdy petla spotka nową liczbę, wrzuca ją do wsplólnego worka

  return counter

sum_list([1,1,1,1,1])
```

```python
#  7 Funkcja count_vowels(text) zliczająca samogłoski w napisie. (ciągi znaków, pętla, in)

def count_vowels(text):
  vowels = 'a,o,e,u,i,y'
  counter = 0
  for x in text:
    if x in vowels:
      counter += 1
  return counter

count_vowels('mateusz')
```

```python
#7.1 oblicz ile jest spolglosek

def count_spgloski(text):
  samogloski = 'a,e,i,o,u,y'
  counter = 0
  for x in text:
    if x not in samogloski:
      counter +=1
  return counter

count_spgloski('joanna')
```

```python
#9.2 znajdz najdluzsze 2
def najdluzsze2(zdanie):
  words = zdanie.split()
  if not words:
    return ""

  longest_word = ""

  for word in words:
    if len(word)> len(longest_word):
      longest_word = word

  return longest_word

print(najdluzsze2("ala nie ma kotaaaa"))
```

```python
#9.2 znajdz najkrotsze 2
#UWAGA! jesli zaczniemy tak jak powyzej, shortest = "", to każde inne bedzie od niego dluzsze

def najkrotsze(zdanie):
  words = zdanie.split()
  if not words: return ""

  najkrotsze = words[0]
  for word in words:
    if len(word) < len(najkrotsze):
      shortest = word

  return shortest

#JESZCZE JEDNO Z CZATA DOKONCZ!
print(najkrotsze("ala ma b"))
```

```python
# 10 Funkcja invert_dict(d) zamieniająca klucze ze wartościami (zakładamy, że wartości są unikalne). (słowniki)

def invert_dict(d):
  new = {} #tworzymy pusty SŁOWNIK
  for key in d.keys(): #iterujemy elementy słownika d
    new[d[key]] = key #nie siegamy po klucz tylko po wartosc klucza, co się staje nowym kluczem
  return new

#d[key]- wartosc to co po dwukropku
d = {'a':1, 'b':2, 'c': 3}

invert_dict(d)
```

```python
# 11 Funkcja count_occurrences(lst) zwracająca słownik liczności elementów listy. (listy, słowniki)

def count_occurrences(lst):
  d = {} #pusty słownik, notes
  for elem in lst:
    if elem in d: #czy mam już ten przedmiot w notesie?
      d[elem] += 1 #mamy+ dodaj sztukę
    else:
      d[elem] = 1 #jeśli nie: wpisz go po raz pierwszy i ustaw liczbę na 1
  return d

count_occurrences([1,1,2,2,3,3,4,4])
```

```python
# 12 Funkcja rekurencyjna factorial(n) obliczająca silnię liczby naturalnej n. (rekurencja)

def factorial(n):
  if n == 0: #rekurencja musi wiedzieć kiedy przestać! bez tego funkcja w nieskonczonosc= 0!=1 wiec to jest bezpiecznik bazowy
    return 1
  else:
    return n * factorial(n-1)
    #musimy n-1 zeby dotrzec do stopu;
    #zeby wyciągnąć to mnożymy n przez silnie poprzedniej wartości (u nas 7) schodzimy ponizej

factorial(8)
```

```python
# 13 Funkcja safe_divide(a, b) zwracająca wynik dzielenia a / b, a w przypadku dzielenia przez zero wypisująca komunikat i zwracająca None. (wyjątki, try/except)

def safe_divide(a, b):
  try:
    return a/b
  except ZeroDivisionError as e:
    print(f"{e}")

safe_divide(6,0)
```
