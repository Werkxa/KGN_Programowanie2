```python
#  1
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

dot = A@B
normA = np.linalg.norm(A)
normB = np.linalg.norm(B)

cos_theta = dot / (normA * normB)
print(cos_theta)
```

```python
# 2
X = np.random.rand(5, 3)
print(X)
print(X.mean(axis=0))
X - X.mean(axis=0)
```

```python
# 3
y_true = np.array([1, 2, 3])
y_pred = np.array([1.1, 1.9, 3.2])

mse = np.mean((y_true - y_pred) ** 2)
print(mse)
a = (y_true - y_pred).mean()
a**2
```

```python
# 4
v = np.array([3, 4])

v_normalized = v / np.linalg.norm(v)
print(v_normalized)

# sprawdzenie
print(np.linalg.norm(v_normalized))

```

```python
# 5
M = np.random.rand(10, 2)
P = np.array([0.5, 0.5])
print(M)
print(P)

distances = np.linalg.norm(M - P, axis=1)
print(M - P)
print(distances)
```

```python
# 6
x = np.array([-2, -1, 0, 1, 2])

relu = np.maximum(0, x)
print(relu * (x > 0))
```

```python
# 7
x = np.array([1, 2,5,5,5,5,5, 2, 3, 100])

mu = np.mean(x)
sigma = np.std(x)

mask = np.abs(x - mu) >= 2 * sigma
outliers = x[mask]

print(outliers)
```
