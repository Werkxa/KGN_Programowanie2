```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 6
rng = np.random.default_rng(42)
n = 200

x = np.linspace(0, 10, n)
noise = rng.normal(0, 1, n)
y = 3 * x + 2 + noise

grupa = rng.choice(["kontrola", "eksperymentalna"], n)

df6 = pd.DataFrame({
    "x": x,
    "noise": noise,
    "y": y,
    "grupa": grupa
})
```

# Heatmap (mapa cieplna)


Wizualizuje wartości w formie kolorów w macierzy. Intensywność koloru odpowiada wielkości wartości.

Zastosowanie:

macierz korelacji,
dane tabelaryczne,
analiza zależności między wieloma zmiennymi.

```python
corr = df6[["x", "y", "noise"]].corr()

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Macierz korelacji")
plt.show()
```

# Histogram 2D

Rozszerzenie histogramu na dwie zmienne – pokazuje zagęszczenie punktów.

Zastosowanie:

analiza skupień,
duże zbiory danych (lepsze niż scatter przy wielu punktach).

```python
plt.hist2d(df6["x"], df6["y"], bins=30)
plt.colorbar()
plt.title("Gęstość punktów")
plt.show()
```

# Area plot (wykres obszarowy)


Wykres liniowy z wypełnieniem pod linią.

Zastosowanie:

zmiany w czasie,
kumulacja wartości,
wizualizacja trendów.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.fill_between(x, y)
plt.title("Wykres obszarowy")
plt.show()
```

# Stem plot (wykres łodygowy)

Pokazuje wartości jako „patyczki” wychodzące z osi.

Zastosowanie:

dane dyskretne,
sygnały,
liczby zdarzeń.

```python
x = np.arange(10)
y = np.random.randint(0, 10, 10)

plt.stem(x, y)
plt.title("Stem plot")
plt.show()
```

# Wykres kołowy — cechy i zastosowania

pokazuje udział procentowy kategorii
suma = 100%
intuicyjny wizualnie

Zastosowania:

struktura populacji

udział segmentów

proporcje klas

```python
# 1
n = 50
rng = np.random.default_rng(42)

wiek = rng.integers(0, 91, n)
plec = rng.choice(["M", "F"], n, [0.6, 0.4])
wzrost = rng.normal(175, 10, n)

waga = 0.4 * wzrost + rng.normal(0, 3, n)

df1 = pd.DataFrame({
    "wiek": wiek,
    "płeć": plec,
    "wzrost_cm": wzrost,
    "waga_kg": waga
})


counts = df1["płeć"].value_counts()

plt.pie(counts, labels=counts.index, autopct="%.1f%%")
plt.title("Struktura płci")
plt.show()

```

# Violin plot — cechy i zastosowania

Cechy:

pokazuje rozkład + gęstość

symetryczny kształt

rozszerzenia = więcej danych

Zastosowania:

porównanie grup

analiza asymetrii

wykrywanie wielomodalności

```python
plt.violinplot(df1["waga_kg"], showmedians=True)
plt.title("Rozkład wagi")
plt.show()
```
