## 2. NumPy – generowanie danych

NumPy służy głównie do pracy na tablicach numerycznych i jest podstawą dla Pandas.

Kluczowe funkcje:

np.array() – tworzy tablicę z listy lub innej struktury.

np.arange(start, stop, step) – generuje ciąg liczb (jak range, ale zwraca tablicę).

np.linspace(start, stop, num) – generuje określoną liczbę punktów w przedziale.

np.random.rand(n, m) – losowe liczby z rozkładu jednostajnego [0,1).

np.random.randn(n, m) – liczby z rozkładu normalnego (średnia 0, wariancja 1).

np.random.randint(low, high, size) – losowe liczby całkowite.

np.random.choice(a, size) – losowy wybór elementów z listy/zbioru.

np.random.seed(42) - ustala stałość generowanych danych losowych


## 3. Pandas – struktury danych i przekształcenia

Pandas operuje na danych tabelarycznych.

Podstawowe obiekty:

pd.Series – jednowymiarowa struktura danych.
pd.DataFrame – tabela (wiersze i kolumny).

Kluczowe funkcje:

pd.DataFrame(data) – tworzy tabelę z danych (np. słownika lub tablicy).

df.head(n) – podgląd pierwszych wierszy.

df.describe() – statystyki opisowe.

df.info() – informacje o typach danych.

df['kolumna'] – dostęp do kolumny.

df.mean(), df.std() – podstawowe statystyki.

pd.concat([df1, df2]) – łączenie danych.

df.apply(func) – zastosowanie funkcji do kolumn.





## 4. Przykład 1 – generowanie prostego zbioru danych

```python
import numpy as np
import pandas as pd

np.random.seed(42)

wiek = np.random.randint(18, 65, 100)
dochód = np.random.normal(5000, 1500, 100)

df = pd.DataFrame({
    "wiek": wiek,
    "dochód": dochód
})

print(df.head())
print(df.describe())
```

## Różne rozkłady pozwalają symulować:

zjawiska naturalne (normalny),

zdarzenia rzadkie (Poissona),

wybory losowe (jednostajny, kategoryczny),

czasy oczekiwania (wykładniczy),

prawdopodobieństwa (beta, binomialny).

## 2. Najważniejsze rozkłady w NumPy (np.random)

### Rozkład jednostajny (uniform)

    np.random.uniform(low=0, high=1, size=100)

wszystkie wartości w przedziale mają równe prawdopodobieństwo
zastosowania: symulacje, losowe próbkowanie


### Rozkład normalny (gaussowski)

    np.random.normal(loc=0, scale=1, size=100)

„dzwonowy” rozkład
wiele zjawisk naturalnych


### Rozkład wykładniczy

    np.random.exponential(scale=1.0, size=100)

modeluje czas między zdarzeniami

wartości dodatnie, asymetryczny


### Rozkład Poissona

    np.random.poisson(lam=3, size=100)

liczba zdarzeń w jednostce czasu/przestrzeni
np. liczba klientów na godzinę

### Rozkład dwumianowy (binomialny)

    np.random.binomial(n=10, p=0.5, size=100)

liczba sukcesów w serii prób
np. liczba „orłów” w 10 rzutach monetą


### Rozkład beta

    np.random.beta(a=2, b=5, size=100)

wartości w przedziale [0,1]
modelowanie prawdopodobieństw


### Rozkład gamma

    np.random.gamma(shape=2.0, scale=2.0, size=100)

uogólnienie wykładniczego
modelowanie czasu życia, opóźnień


### Rozkład log-normalny

    np.random.lognormal(mean=0, sigma=1, size=100)

dane dodatnie, silnie skośne
np. dochody, ceny


### Rozkład kategoryczny (losowanie z listy)

    np.random.choice(["A", "B", "C"], size=100, p=[0.2, 0.5, 0.3])

dane jakościowe
można ustawić prawdopodobieństwa

Uwagi: klasyczny przypadek do analizy korelacji i regresji.

```python
import numpy as np
import pandas as pd


```

```python
# 1

n = 50
rng = np.random.default_rng(42)


wiek = rng.integers(0, 91, n)
plec = rng.choice(["M", "F"], n, [0.6, 0.4])
wzrost = rng.normal(175, 10, n)

# zależność: waga częściowo zależy od wzrostu
waga = 0.4 * wzrost + rng.normal(0, 3, n)

df1 = pd.DataFrame({
    "wiek": wiek,
    "płeć": plec,
    "wzrost_cm": wzrost,
    "waga_kg": waga
})

df1.head()
```

```python
# 2
rng = np.random.default_rng(42)
n = 50

wizyty = rng.poisson(5, n)
czas = rng.exponential(10, n)
wydatki = rng.lognormal(2, 1, n)
segment = rng.choice(["A", "B", "C"], n, p=[0.3, 0.5, 0.2])

df2 = pd.DataFrame({
    "liczba_wizyt": wizyty,
    "czas_na_stronie": czas,
    "wydatki": wydatki,
    "segment": segment
})

df2.head()
```

```python
# 3
rng = np.random.default_rng(43)
n = 50


wynik = np.clip(rng.normal(60, 15, n), 0, 100)

czas = rng.normal(60, 10, n)
zaliczony = (wynik > 50).astype(int)
grupa = rng.choice(["A", "B"], n)

df3 = pd.DataFrame({
    "wynik": wynik,
    "czas_rozwiązywania": czas,
    "zaliczony": zaliczony,
    "grupa": grupa
})

df3.head()

```

```python
#  4
rng = np.random.default_rng(42)
n = 50

temp = rng.normal(20, 5, n)
wilg = rng.uniform(30, 90, n)
zdarzenia = rng.poisson(2, n)
czas_miedzy = rng.exponential(5, n)

df4 = pd.DataFrame({
    "temperatura": temp,
    "wilgotność": wilg,
    "liczba_zdarzeń": zdarzenia,
    "czas_między_zdarzeniami": czas_miedzy
})

df4.head()

df4.describe()
```

```python
# 5
rng = np.random.default_rng(42)
n = 50

dochód = rng.lognormal(8, 0.7, n)
wydatki = 0.7 * dochód + rng.normal(0, 1000, n)
oszczędności = dochód - wydatki

klasa = rng.choice(
    ["niska", "średnia", "wysoka"],
    n,
    p=[0.5, 0.4, 0.1]
)

df5 = pd.DataFrame({
    "dochód": dochód,
    "wydatki": wydatki,
    "oszczędności": oszczędności,
    "klasa_społeczna": klasa
})

df5.head()
```

```python
# 6
rng = np.random.default_rng(42)
n = 200

x = np.linspace(0, 10, n)
noise = rng.normal(0, 1, n)
y = 3 * x + 2 + noise

grupa = rng.choice(["kontrola", "eksperymentalna"], n)

df6 = pd.DataFrame({
    "x": x,
    "noise": noise,
    "y": y,
    "grupa": grupa
})

df6.head()
```
