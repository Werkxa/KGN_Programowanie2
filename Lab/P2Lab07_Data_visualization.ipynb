```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rng = np.random.default_rng(42)
data = rng.normal(0, 1, 1000)

plt.hist(data, bins=30)
plt.title("Histogram – rozkład normalny")
plt.xlabel("wartość")
plt.ylabel("liczba obserwacji")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = 3*x + 2 + rng.normal(0, 1, 100)

plt.scatter(x, y)
plt.title("Zależność liniowa")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.plot(x, y)
plt.title("Wykres liniowy")
plt.show()
```

```python
kategorie = ["A", "B", "C"]
wartosci = [10, 20, 15]

plt.bar(kategorie, wartosci)
plt.title("Wykres słupkowy")
plt.show()
```

```python
data = rng.normal(0, 1, 100)

plt.boxplot(data)
plt.title("Boxplot")
plt.show()
```

# ZADANIA

Przygotuj po 2 wykresy dla każdego zadania z poprzedniego tygodnia



```python
# 1
n = 50
rng = np.random.default_rng(42)


wiek = rng.integers(0, 91, n)
plec = rng.choice(["M", "F"], n, [0.6, 0.4])
wzrost = rng.normal(175, 10, n)

# zależność: waga częściowo zależy od wzrostu
waga = 0.4 * wzrost + rng.normal(0, 3, n)

df1 = pd.DataFrame({
    "wiek": wiek,
    "płeć": plec,
    "wzrost_cm": wzrost,
    "waga_kg": waga
})

plt.scatter(df1["wzrost_cm"], df1["waga_kg"])
plt.show()
```
