3. Atrybuty i Kształt (Shape)
Zrozumienie wymiarów jest kluczowe w algebrze liniowej.

    A = np.array([[1, 2, 3], [4, 5, 6]])

    print(f"Wymiary (ndim): {A.ndim}")
    print(f"Kształt (shape): {A.shape}") # (wiersze, kolumny)
    print(f"Typ danych: {A.dtype}")

# Zmiana kształtu (Reshaping)

    B = np.arange(12).reshape(3, 4)
    print(B)

4. Operacje matematyczne (Element-wise)

W NumPy operacje są wykonywane na każdym elemencie z osobna (tzw. wektoryzacja).

    x = np.array([1, 2, 3])
    y = np.array([4, 5, 6])

    print(x + y)  # [5, 7, 9]
    print(x * 2)  # [2, 4, 6]
    print(x ** 2) # [1, 4, 9]
    print(np.sin(x))

5. Algebra Liniowa

Używamy modułu np.linalg lub dedykowanych operatorów.

Mnożenie macierzy
Pamiętaj: operacja * to mnożenie element po elemencie. Aby pomnożyć macierze zgodnie z zasadami algebry, używamy @ lub .dot().


    A = np.array([[1, 2], [3, 4]])
    B = np.array([[5, 6], [7, 8]])

## Mnożenie macierzowe

    C = A @ B

Lub:

    C = np.dot(A, B)

# Transpozycja

    print(A.T)

# Wyznacznik macierzy (Determinant)

    det = np.linalg.det(A)

# Macierz odwrotna

    A_inv = np.linalg.inv(A)

# Rozwiązywanie układu równań Ax = b

    1x + 2y = 5
    3x + 4y = 11

    b = np.array([5, 11])

    rozwiazanie = np.linalg.solve(A, b)
    print(f"Wynik układu równań: {rozwiazanie}")


6. Agregacja danych
Szybkie statystyki na dużych zbiorach danych.


    data = np.array([[1, 2], [3, 4]])

    print(np.mean(data, axis=0)) # Średnia z kolumn

    print(np.max(data, axis=1))  # Maksimum z wierszy
