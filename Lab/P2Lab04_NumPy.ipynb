4. Operacje matematyczne (Element-wise)

W NumPy operacje są wykonywane na każdym elemencie z osobna (tzw. wektoryzacja).

    x = np.array([1, 2, 3])
    y = np.array([4, 5, 6])

    print(x + y)  # [5, 7, 9]
    print(x * 2)  # [2, 4, 6]
    print(x ** 2) # [1, 4, 9]
    print(np.sin(x))

### Zadanie 8: Normalizacja Min-Max
Przeskaluj wektor [10, 20, 30, 40, 50] do zakresu 0-1.

$$v_{new} = \frac{v - min(v)}{max(v) - min(v)}$$

```python
v = np.array([10, 20, 30, 40, 50])
v_min = v.min()
v_max = v.max()

v_norm = (v - v_min) / (v_max - v_min)
v_norm


```

```python
import numpy as np

wyniki = np.array([20, 55, 80, 45, 90, 70])

# 1. Definiujemy listę warunków
warunki = [
    (wyniki < 50),
    (wyniki >= 50) & (wyniki <= 75),
    (wyniki > 75)
]

# 2. Definiujemy listę wartości przypisanych do warunków
wybory = ["Niedostateczny", "Dobry", "Bardzo dobry"]

# 3. Wykonujemy operację (default to wartość, gdy żaden warunek nie pasuje)
oceny = np.select(warunki, wybory, default="Błąd")

print(oceny)
# Wynik: ['Niedostateczny' 'Dobry' 'Bardzo dobry' 'Niedostateczny' 'Bardzo dobry' 'Dobry']
```
